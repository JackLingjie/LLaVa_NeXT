```python
import yaml  
import json  
  
def load_yaml(file_path):  
    with open(file_path, 'r') as file:  
        return yaml.safe_load(file)  
  
def extract_json_paths_and_strategies(data):  
    return {dataset['json_path']: dataset['sampling_strategy'] for dataset in data['datasets']}  
  
# 加载 YAML 文件  
single_image_data = load_yaml('/home/v-lingjiang/project/LLaVa_NeXT/scripts/train/single_image.yaml')  
onevision_data = load_yaml('/home/v-lingjiang/project/LLaVa_NeXT/scripts/train/onevision.yaml')  
  
# 提取 json_path 和 sampling_strategy 字典  
single_image_dict = extract_json_paths_and_strategies(single_image_data)  
onevision_dict = extract_json_paths_and_strategies(onevision_data)  
  
# 找出重叠的 json_path 及其在 onevision.yaml 中的 sampling_strategy  
common_items = [  
    {'json_path': json_path, 'sampling_strategy': onevision_dict[json_path]}  
    for json_path in single_image_dict  
    if json_path in onevision_dict  
]  
  
# # 将结果写入 JSON 文件  
with open('onevision_single_image.json', 'w') as json_file:  
    json.dump(common_items, json_file, indent=4)  
  
print("Matching json_path and their sampling_strategy from onevision.yaml have been saved to common_paths.json.")  
```
