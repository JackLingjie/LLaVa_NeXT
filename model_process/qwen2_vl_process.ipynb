```python
import torch
from transformers import (
    Qwen2VLForConditionalGeneration,  # Qwen2-VL
    Qwen2ForCausalLM,                 # Qwen2 / Qwen2.5 coder
)

# 1) 加载 Qwen2-VL 模型
qwen2vl_model_path = "/mnt/lingjiejiang/textual_aesthetics/model_checkpoint/vlm_checkpoints/Qwen2-VL-7B-Instruct"
qwen2vl = Qwen2VLForConditionalGeneration.from_pretrained(
    qwen2vl_model_path,
    torch_dtype="auto",  
    device_map="auto"
)

# 2) 加载 Qwen2.5 coder 模型 (或者 Qwen2-7B-Instruct 等)
qwen2coder_model_path = "/mnt/lingjiejiang/multimodal_code/checkpoints/llms/Qwen2.5-Coder-7B-Instruct"
qwen2coder = Qwen2ForCausalLM.from_pretrained(
    qwen2coder_model_path,
    torch_dtype="auto",
    device_map="auto"
)

# 取出 Qwen2VL 里 “LLM 部分”的 state_dict
qwen2vl_backbone_sd = qwen2vl.model.state_dict()

# 取出 Qwen2Coder (Qwen2.5) 里 “LLM 部分”的 state_dict
qwen2coder_sd = qwen2coder.model.state_dict()

# 遍历 coder 的每个参数，如果在 vl 里有同名参数且形状一致，就直接覆盖
for name, param in qwen2coder_sd.items():
    if name in qwen2vl_backbone_sd and qwen2vl_backbone_sd[name].shape == param.shape:
        # 用 coder 的参数覆盖 VL 的对应参数
        with torch.no_grad():
            qwen2vl_backbone_sd[name].copy_(param)

# 将更新后的 state_dict load 回去
qwen2vl.model.load_state_dict(qwen2vl_backbone_sd)

qwen2vl_lm_head_sd = qwen2vl.lm_head.state_dict()
qwen2coder_lm_head_sd = qwen2coder.lm_head.state_dict()
for name, param in qwen2coder_lm_head_sd.items():
    if name in qwen2vl_lm_head_sd and qwen2vl_lm_head_sd[name].shape == param.shape:
        with torch.no_grad():
            qwen2vl_lm_head_sd[name].copy_(param)
qwen2vl.lm_head.load_state_dict(qwen2vl_lm_head_sd)

qwen2vl.save_pretrained("/mnt/lingjiejiang/multimodal_code/checkpoints/vlms/Qwenvl2-coder2.5-7B-Instruct_merge")
```

```python
from transformers import AutoTokenizer  
  
# 加载 Qwen2Coder 的 tokenizer  
qwen2coder_tokenizer_path = "/mnt/lingjiejiang/multimodal_code/checkpoints/llms/Qwen2.5-Coder-7B-Instruct"  
qwen2coder_tokenizer = AutoTokenizer.from_pretrained(qwen2coder_tokenizer_path)  
  
# 加载 Qwen2VL 的 tokenizer  
qwen2vl_tokenizer_path = "/mnt/lingjiejiang/textual_aesthetics/model_checkpoint/vlm_checkpoints/Qwen2-VL-7B-Instruct"  
qwen2vl_tokenizer = AutoTokenizer.from_pretrained(qwen2vl_tokenizer_path)  
  
# 获取两个词表的词汇和 ID  
qwen2coder_vocab = qwen2coder_tokenizer.get_vocab()  
qwen2vl_vocab = qwen2vl_tokenizer.get_vocab()  
  
# 反转字典以便通过 ID 获取 token  
qwen2coder_id_to_token = {v: k for k, v in qwen2coder_vocab.items()}  
  
# 找到需要替换和添加的 token  
tokens_to_replace = {}  
tokens_to_add = {}  
  
# 查找需要替换的 token  
for token, vl_id in qwen2vl_vocab.items():  
    if token in qwen2coder_vocab:  
        coder_id = qwen2coder_vocab[token]  
        if vl_id != coder_id:  
            tokens_to_replace[vl_id] = qwen2coder_id_to_token[coder_id]  
    else:  
        tokens_to_add[token] = vl_id  
  
# 添加 qwen2coder 中有但 qwen2vl 中没有的 token，直接使用 coder 的 ID  
for token, coder_id in qwen2coder_vocab.items():  
    if token not in qwen2vl_vocab:  
        tokens_to_add[token] = coder_id  
  
# # 更新 qwen2vl 的词表  
updated_vocab = qwen2vl_vocab.copy()  
for vl_id, coder_token in tokens_to_replace.items():  
    updated_vocab[coder_token] = vl_id  
  
for token, coder_id in tokens_to_add.items():  
    updated_vocab[token] = coder_id  
  
# # 创建新的 tokenizer  
# updated_tokenizer = qwen2vl_tokenizer  
# updated_tokenizer.vocab = updated_vocab  
# updated_tokenizer.ids_to_tokens = {v: k for k, v in updated_vocab.items()}  
  
# # 保存更新后的 tokenizer  
# updated_tokenizer.save_pretrained("/mnt/lingjiejiang/multimodal_code/checkpoints/vlms/Qwenvl2-coder2.5-7B-Instruct_updated")  
```
